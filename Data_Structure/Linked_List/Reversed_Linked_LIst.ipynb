```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
# Solution 1: Iterative using two pointers curr and prev, curr to head and shift over head, prev get earlier curr
# T O(n), M O(1)
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # Base case: if the head is None or there's only one node in the list
        prev, curr = None, head
        while curr:    
            nxt = curr.next
            curr.next = prev     
            prev = curr
            curr = nxt
        return prev
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
# Solution 2: Recursive by breaking down problem into sub problems, 1 -> 2 -> 3 into 2 -> 3 and into 3 -> None reverse
# T O(n), M O(n)

class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # Base case: if the head is None or there's only one node in the list
        if not head:
            return None
        new_head = head    
        if head.next:            
            new_head = self.reverseList(head.next)
            head.next.next = head
        head.next = None 
            
        return new_head
```
