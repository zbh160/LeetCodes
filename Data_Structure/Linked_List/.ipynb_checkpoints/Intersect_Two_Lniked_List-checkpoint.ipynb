```python
# Reference https://leetcode.com/problems/intersection-of-two-linked-lists/

# Logic 

# If one of the list gets end, return the other 
# Start with the first node of 2 lists
# Compare value of nodes
# If condition mathces, go with the list selected and select next node
# e.g. l1: 1 1 2 2 3 4 ; l2: 1 3 4, order should be l1 1 -> l1 1 -> l2 1 -> l1 2 -> l1 2 -> l1 3 -> l2 3 -> l1 4 -> l2 4
```

```python
# Definition for singly-linked list.

'''
    class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        A, B = headA, headB
        while A != B:
            A = A.next if A else headB
            B = B.next if B else headA
        return A
'''

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        l1 = headA
        l2 = headB
        print("Start Node: ")
        print(l1.data)
        print(l2.data)
        i = 0
        while (l1 != l2):
            i = i + 1
            print("Step %i"%i)
            print("New Node:")
            if l1 is None:
                l1 = headB
                print("end of l1: %s"%l1.data)
            else:
                if l1.next is None:
                    l1 = headB
                    print("end of l1: %s"%l1.data)
                else:
                    l1 = l1.next
                    print(l1.data)
                    
            if l2 is None:
                l2 = headA
                print("end of l2: %s"%l2.data)
            else:
                if l2.next is None:
                    l2 = headA
                    print("end of l2: %s"%l2.data)
                else:
                    l2 = l2.next
                    print(l2.data)

            # l1 = headB if not l1 else l1.next
            # l2 = headA if not l2 else l2.next
        
        if l1 is not None:
            print("%i Steps taken, intersection found: %s"%(i,l1.data))
        else:
            print("%i Steps taken, no intersection found"%i)
        return l1

```

```python
# Python program to get intersection point of two linked list
# Two Loops to find it

# Link list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
 
# function to get the intersection point of two linked lists head1 and head
def getIntersectionNode(head1, head2):
    while head2:
        temp = head1
        while temp:
            # if both Nodes are same
            if temp == head2:
                return head2
            temp = temp.next
        head2 = head2.next
    # intersection is not present between the lists
    return None
 
 
# Driver Code
if __name__ == '__main__':
    '''
    Create two linked lists
 
    1st 3->6->9->15->30
    2nd 10->15->30
 
    15 is the intersection point
    '''
 
    newNode = Node(10)
    head1 = newNode
    newNode = Node(3)
    head2 = newNode
    newNode = Node(6)
    head2.next = newNode
    newNode = Node(9)
    head2.next.next = newNode
    newNode = Node(15)
    head1.next = newNode
    head2.next.next.next = newNode
    newNode = Node(30)
    head1.next.next = newNode
 
    intersectionPoint = getIntersectionNode(head1, head2)
 
    if not intersectionPoint:
        print(" No Intersection Point ")
    else:
        print("Intersection Point:", intersectionPoint.data)
 
```
