```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    Front = None
    def isPalindrome(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        slow = head
        fast = head
        # step 1 get middle element
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
        # step 2 if no. of nodes are odd then move slow to one point
        if fast is not None and fast.next is None:
            slow = slow.next
        # step 3 Reverse the end half
        prev = None
        temp = None
        while slow is not None and slow.next is not None:
            temp = slow.next
            slow.next = prev
            prev = slow
            slow = temp
        if slow is not None:
            slow.next = prev
        # step 4 compare the start half and end half if found any inequality 
        fast = head
        while slow and fast:
            if slow.val != fast.val:
                return False
            slow = slow.next
            fast = fast.next
        return True
```
