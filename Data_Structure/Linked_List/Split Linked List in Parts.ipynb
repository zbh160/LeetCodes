```python
# Reference https://leetcode.com/problems/split-linked-list-in-parts/description/
# Logic 
# Get length of list using loop, remains and default value
# Assign array of int with num of nodes assgined to each part, remain will be added evenly to each part using loop
# traverse the list starting from head, while head is not None, then loop over num of nodes in each part 
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def splitListToParts(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: List[ListNode]
        """
        # get length l
        # each array should be l // k, mod l % k would be remain to be added to each part
        # double loop over each part to increment until mod gets 0 
        # create k new list inside result


        # use while head:
        # for i in k range
        # add head.val, head = head.next
        # if head is None, then add None to list until k
        # if k but head.next is not None, do another round k
        # not working, because each part has to be in order 
        result = []
       
        
        l = 0
        start = head
        while head:
            l += 1
            head = head.next
        default = l // k
        remain = l % k
        i = 0
        arrange = []
        for i in range(k):
            if remain :
                arrange.append(default + 1)  # add 1 till the reaminder become 0
                remain -= 1
            else:
                arrange.append(default)
        # print(arrange)
        # now traverse the linked list and split the node 
        j = 0
        ptr = start
        head = start
        i = 0
        while(ptr) :
            q = ptr 
            i += 1
            ptr = ptr.next 
            # if the new length i is equal to the arrange[i] then just make that nodes next pointer to None
            if i == arrange[j] :
                # Cut the list to head to q, point q.next to None
                q.next = None
                result.append(head)
                head = ptr 
                i = 0
                j += 1
        # print(j)
        #print(result)
        # if the length is not satisfied then just add none
        for i in range(j,k):
            result.append(None)
            # result.append(temp)
        
        return result
```
